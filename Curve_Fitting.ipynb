曲线拟合（Curve Fitting）是通过数学方法找到一条曲线，使其尽可能好地描述一组数据点之间的关系。它是数据分析和科学计算中的重要技术，用于：

- 发现变量间的数学关系
- 预测未知数据
- 理解数据背后的规律
- 去除噪声影响

主要方法
1. 线性回归
最简单的曲线拟合形式，寻找最佳拟合直线。

2. 多项式回归
使用多项式函数（二次、三次等）拟合非线性数据。

3. 非线性回归
使用指数、对数、三角函数等复杂函数进行拟合。

线性拟合（Linear Fitting）

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import seaborn as sns

sns.set_style("dark")

# 设置中文字体和图形样式
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

```

```python
# 生成带噪声的线性数据
np.random.seed(42)
x = np.linspace(0, 10, 50)
y_true = 2 * x + 1
y = y_true + np.random.normal(0, 2, size=x.shape)

# 线性拟合
coefficients = np.polyfit(x, y, 1)  # 1表示一次多项式
y_fit = np.polyval(coefficients, x)

# 绘图
plt.figure(figsize=(10, 6))
plt.scatter(x, y, alpha=0.6, label='数据点', color='blue')
plt.plot(x, y_fit, 'r-', linewidth=2, label=f'拟合直线: y = {coefficients[0]:.2f}x + {coefficients[1]:.2f}')
plt.plot(x, y_true, 'g--', alpha=0.7, label='真实关系: y = 2x + 1')
plt.xlabel('x')
plt.ylabel('y')
plt.title('线性曲线拟合示例')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print(f"拟合参数: 斜率 = {coefficients[0]:.3f}, 截距 = {coefficients[1]:.3f}")
```
